**Manejo de Retiro y Asistencia**

 Se debe identificar a los estudiantes que se han retirado (celdas vacas) o que no han asistido (celdas con valor cero).

```python
# Primero, gestionamos los datos de asistencia y retiro
eval_columns = ['EVALUACIN_1', 'EVALUACIN_2', 'EVALUACIN_3', 'EVALUACIN_4']
df[eval_columns] = df[eval_columns].apply(pd.to_numeric, errors='coerce')

# Crear la columna 'Retirado' si todas las evaluaciones estn vacas (NaN)
df['Retirado'] = df[eval_columns].isna().all(axis=1)

# Crear la columna 'No_Asisti' para aquellos estudiantes que tienen valor 0 en cualquier evaluacin
df['No_Asisti'] = (df[eval_columns] == 0).any(axis=1)

# Reemplazar los NaN con 0 en las evaluaciones para poder calcular la nota final
df[eval_columns] = df[eval_columns].fillna(0)

```

**Clculo de la Nota Final**

Despus de manejar el retiro y la asistencia, se calcula la nota final usando los pesos proporcionados para cada evaluacin.

```python
# Calcular la nota final basada en los pesos de las evaluaciones
df['Nota_Final'] = (df['EVALUACIN_1'] * 0.15 +
                    df['EVALUACIN_2'] * 0.20 +
                    df['EVALUACIN_3'] * 0.30 +
                    df['EVALUACIN_4'] * 0.35)

# Determinar si el estudiante est aprobado o no (nota mayor a 12.5)
df['Aprobado'] = df['Nota_Final'] >= 12.5

# Modificar la condicin para determinar si el estudiante est desaprobado (nota mayor a 0 y menor a 12.5)
df['Desaprobado'] = (df['Nota_Final'] > 0) & (df['Nota_Final'] < 12.5)

# Mantener la columna 'Retirado' o 'No Asisti' segn corresponda (asumo que ya existen)

```

```python
# Reaplicar los filtros con los valores correctos
aprobados_students = df[df['Aprobado'] == True]
desaprobados_students = df[df['Desaprobado'] == True]
retirados_students = df[df['Retirado'] == True]
inasistentes_students = df[df['No_Asisti'] == True]

# Contar el nmero de estudiantes en cada categora para verificar si los filtros funcionan correctamente
print("Aprobados:", aprobados_students.shape[0])
print("Desaprobados:", desaprobados_students.shape[0])
print("Retirados:", retirados_students.shape[0])
print("No asistieron:", inasistentes_students.shape[0])
```

```python
import numpy as np

# Funcin para detectar outliers usando el mtodo del IQR
def detectar_anomalias(df, columna):
    if pd.api.types.is_numeric_dtype(df[columna]):  # Verificar si la columna es numrica
        Q1 = df[columna].quantile(0.25)
        Q3 = df[columna].quantile(0.75)
        IQR = Q3 - Q1
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR

        # Detectar valores fuera de los lmites
        anomalias = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]
        return anomalias
    else:
        return pd.DataFrame()  # Si la columna no es numrica, retornar un DataFrame vaco

# Variables predictoras (asegrate de incluir solo las numricas o usar la funcin anterior para filtrar)
variables = ['EVALUACIN_1', 'TURNO', 'PROGRAMA', 'SEDE_ESTUDIANTE', 'AREA']

# Detectar anomalias para cada variable
for variable in variables:
    anomalias = detectar_anomalias(df, variable)
    if not anomalias.empty:  # Verificar si hay anomalas detectadas
        print(f"En la variable {variable}, se encontraron {anomalias.shape[0]} valores anmalos.")
    else:
        print(f"La variable {variable} no es numrica o no tiene valores anmalos detectados.")

```

```python
# Diccionario para almacenar los outliers por sede
outliers = {}

# Create a sample DataFrame for demonstration purposes - replace this with your actual data loading
import pandas as pd
data = {'SEDE_ESTUDIANTE': ['Sede1', 'Sede2', 'Sede1', 'Sede3', 'Sede2'],
        'Nota_Final': [80, 90, 75, 85, 95]}
data_filtered = pd.DataFrame(data)

# Iterar sobre cada sede nica en el conjunto de datos filtrado
for sede in data_filtered['SEDE_ESTUDIANTE'].unique():
    # Filtrar los datos de la sede actual
    sede_data = data_filtered[data_filtered['SEDE_ESTUDIANTE'] == sede]['Nota_Final']

    # Calcular el primer y tercer cuartil
    q1 = sede_data.quantile(0.25)
    q3 = sede_data.quantile(0.75)

    # Calcular el rango intercuartlico (IQR)
    iqr = q3 - q1

    # Definir la condicin de outliers (valores fuera de 1.5 * IQR por debajo de Q1 o por encima de Q3)
    outlier_condition = ((sede_data < (q1 - 1.5 * iqr)) | (sede_data > (q3 + 1.5 * iqr)))

    # Almacenar los outliers en el diccionario por sede
    outliers[sede] = sede_data[outlier_condition]

# Mostrar valores atpicos por sede con mejor visualizacin
for sede, values in outliers.items():
    # Convertir los valores a una lista redondeada a dos decimales
    rounded_values = [round(value, 2) for value in values]

    # Formatear la salida con saltos de lnea cada 10 valores
    formatted_values = '\n'.join([', '.join(map(str, rounded_values[i:i+10])) for i in range(0, len(rounded_values), 10)])

    # Imprimir los resultados de outliers por sede
    print(f'{sede} outliers:\n{formatted_values}\n')
```

# Preprosesamiento

## Crear las clases objetivo 'target_classes'

```python
# 1. Retirado: No tiene ninguna evaluacin registrada
df['Retirado'] = df[['EVALUACIN_1', 'EVALUACIN_2', 'EVALUACIN_3', 'EVALUACIN_4']].isna().all(axis=1)

# 2. No Asisti: Tiene alguna evaluacin con un valor de 0 (no asisti)
df['No_Asisti'] = (df[['EVALUACIN_1', 'EVALUACIN_2', 'EVALUACIN_3', 'EVALUACIN_4']] == 0).any(axis=1)

# 3. Aprobado o Desaprobado
# Nota >= 12.5 -> Aprobado
df['Aprobado'] = np.where(df['Nota_Final'] >= 12.5, 1, 0)

# Crear la columna 'target_classes' para las clases objetivo
# Retirado = 3, No Asisti = 2, Aprobado = 1, Desaprobado = 0
df['target_classes'] = np.select(
    [
        df['Retirado'] == True,  # Retirado
        df['No_Asisti'] == True,  # No Asisti
        df['Aprobado'] == 1  # Aprobado
    ],
    [
        3,  # Retirado
        2,  # No Asisti
        1   # Aprobado
    ],
    default=0  # Desaprobado
)

# Verificar la distribucin de las clases objetivo
print(df['target_classes'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

# Identificar las columnas categricas que an no estn codificadas
categorical_columns = ['PROGRAMA', 'TURNO','AREA','SEDE_ESTUDIANTE',]

# Crear un codificador
label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col].astype(str))

    # Para las etiquetas no vistas, asigna un valor predeterminado (ej. -1)
    def handle_unseen_labels(value):
        if value not in le.classes_:
            return -1  # O algn valor que represente etiquetas no vistas
        return le.transform([value])[0]

    # Aplicar la transformacin a X_test asegurando manejar etiquetas no vistas
    X_test[col] = X_test[col].apply(lambda x: handle_unseen_labels(x))

    label_encoders[col] = le  # Guardar el codificador en caso sea necesario para revertir luego

```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score

def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    # Entrenar el modelo
    model.fit(X_train, y_train)

    # Predecir en el conjunto de entrenamiento y de prueba
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Calcular mtricas de precisin
    accuracy_train = accuracy_score(y_train, y_train_pred)
    accuracy_test = accuracy_score(y_test, y_test_pred)

    return {
        'Modelo': model_name,
        'Accuracy (Train)': round(accuracy_train, 4),
        'Accuracy (Test)': round(accuracy_test, 4)
    }

# Definir modelos a evaluar
modelos = {
    'Random Forest': RandomForestClassifier(),
    'XGBoost (Binary Logistic)': xgb.XGBClassifier(
    objective='binary:logistic',  # Objetivo de clasificacin binaria
    n_estimators=50,  # Nmero de rboles
    learning_rate=0.5,  # Tasa de aprendizaje
    seed=1,  # Control de aleatoriedad
    random_state=42  # Para asegurar reproducibilidad
),
    'XGBoost (Multiclass Softmax)': XGBClassifier(objective="multi:softmax",
                                  num_class=3,
                                  n_estimators=50,  # Nmero de rboles
                                  learning_rate=1,  # Tasa de aprendizaje
                                  seed=1,  # Control de la aleatoriedad (equivalente a random_state)
                                  random_state=42),
    'CatBoost': CatBoostClassifier(
    iterations=100,        # Nmero de iteraciones (puedes ajustar este valor segn el tiempo y recursos disponibles)
    depth=6,               # Profundidad de los rboles
    learning_rate=0.5,     # Tasa de aprendizaje
    loss_function='MultiClass',  # Para problemas de clasificacin multiclase
    random_seed=42,        # Para reproducibilidad
    verbose=50             # Imprime el progreso cada 50 iteraciones
)
}

resultados = []

# Entrenar y evaluar cada modelo
for nombre_modelo, modelo in modelos.items():
    result = evaluate_model(modelo, X_train, y_train, X_test, y_test, nombre_modelo)
    resultados.append(result)



# Crear DataFrame y mostrar resultados
df_resultados = pd.DataFrame(resultados)
print(df_resultados[['Modelo', 'Accuracy (Train)', 'Accuracy (Test)']])
```

```python
import gradio as gr
import joblib
import pandas as pd

# Cargar los modelos previamente entrenados
rf_model = joblib.load('modelo_RandomForestClassifier.pkl')
xgb_model_softmax = joblib.load('modelo_XGBClassifier.pkl')
modelo = joblib.load('modelo_xgb.pkl')
modelo_catboost = joblib.load('modelo_catboost.pkl')

# Cargar los encoders
label_encoder_sede = joblib.load('label_encoder_sede_estudiante.pkl')
label_encoder_programa = joblib.load('label_encoder_programa.pkl')
label_encoder_area = joblib.load('label_encoder_area.pkl')
label_encoder_turno = joblib.load('label_encoder_turno.pkl')

# Funcin de prediccin con mltiples modelos
def predict(sede, programa, area, turno, evaluacion1):
    try:
        # Validacin de entradas
        if evaluacion1 < 0 or evaluacion1 > 20:
            return "Error: La evaluacin debe estar entre 0 y 20"

        # Transformar las entradas usando los encoders
        sede_encoded = label_encoder_sede.transform([sede])[0]
        programa_encoded = label_encoder_programa.transform([programa])[0]
        area_encoded = label_encoder_area.transform([area])[0]
        turno_encoded = label_encoder_turno.transform([turno])[0]

        # Preparar los datos para la prediccin
        input_data = [[sede_encoded, programa_encoded, area_encoded, turno_encoded, evaluacion1]]

        # Realizar la prediccin usando todos los modelos (aprobacin, desaprobacin e inasistencia)
        rf_prob_approval = rf_model.predict_proba(input_data)[0, 1] * 100
        rf_prob_inasistencia = rf_model.predict_proba(input_data)[0, 2] * 100
        rf_prob_desaprobacion = rf_model.predict_proba(input_data)[0, 0] * 100

        xgb_prob_approval = modelo.predict_proba(input_data)[0, 1] * 100
        xgb_prob_inasistencia = modelo.predict_proba(input_data)[0, 2] * 100
        xgb_prob_desaprobacion = modelo.predict_proba(input_data)[0, 0] * 100

        xgb_softmax_approval = xgb_model_softmax.predict_proba(input_data)[0, 1] * 100
        xgb_softmax_inasistencia = xgb_model_softmax.predict_proba(input_data)[0, 2] * 100
        xgb_softmax_desaprobacion = xgb_model_softmax.predict_proba(input_data)[0, 0] * 100

        catboost_prob_approval = modelo_catboost.predict_proba(input_data)[0, 1] * 100
        catboost_prob_inasistencia = modelo_catboost.predict_proba(input_data)[0, 2] * 100
        catboost_prob_desaprobacion = modelo_catboost.predict_proba(input_data)[0, 0] * 100

        # Crear el resultado agrupado en cuadros separados para aprobacin, desaprobacin e inasistencia
        approval_result = (f'<div style="border:1px solid #233269; padding:10px; margin-bottom:10px;">'
                           f'<h2 style="color: #233269;">Probabilidades de Aprobacin</h2>'
                           f'<ul>'
                           f'<li>Probabilidad de aprobacin (Random Forest): {rf_prob_approval:.2f}%</li>'
                           f'<li style="background-color: #feddb5; padding:5px;">Probabilidad de aprobacin (XGBoost - Binary): {xgb_prob_approval:.2f}%</li>'  # Sombreado aqu
                           f'<li>Probabilidad de aprobacin (XGBoost - Softmax): {xgb_softmax_approval:.2f}%</li>'
                           f'<li>Probabilidad de aprobacin (CatBoost): {catboost_prob_approval:.2f}%</span></li>'
                           f'</ul>'
                           f'</div>')

        desaprobacion_result = (f'<div style="border:1px solid #ed4255; padding:10px; margin-bottom:10px;">'
                                f'<h2 style="color: #ed4255;">Probabilidades de Desaprobacin</h2>'
                                f'<ul>'
                                f'<li>Probabilidad de desaprobacin (Random Forest): {rf_prob_desaprobacion:.2f}%</li>'
                                f'<li style="background-color: #feddb5; padding:5px;">Probabilidad de desaprobacin (XGBoost - Binary): {xgb_prob_desaprobacion:.2f}%</li>'  # Sombreado aqu
                                f'<li>Probabilidad de desaprobacin (XGBoost - Softmax): {xgb_softmax_desaprobacion:.2f}%</li>'
                                f'<li>Probabilidad de desaprobacin (CatBoost): {catboost_prob_desaprobacion:.2f}%</span></li>'
                                f'</ul>'
                                f'</div>')

        inasistencia_result = (f'<div style="border:1px solid #ad4992; padding:10px; margin-bottom:10px;">'
                               f'<h2 style="color: #ad4992;">Probabilidades de Inasistencia</h2>'
                               f'<ul>'
                               f'<li>Probabilidad de inasistencia (Random Forest): {rf_prob_inasistencia:.2f}%</li>'
                               f'<li style="background-color: #feddb5; padding:5px;">Probabilidad de inasistencia (XGBoost - Binary): {xgb_prob_inasistencia:.2f}%</li>'  # Sombreado aqu
                               f'<li>Probabilidad de inasistencia (XGBoost - Softmax): {xgb_softmax_inasistencia:.2f}%</li>'
                               f'<li>Probabilidad de inasistencia (CatBoost): {catboost_prob_inasistencia:.2f}%</span></li>'
                               f'</ul>'
                               f'</div>')

        # Combinar los resultados en el orden solicitado: aprobacin, desaprobacin, inasistencia
        return approval_result + desaprobacion_result + inasistencia_result

    except Exception as e:
        return f"Ocurri un error: {str(e)}"

# Encabezado HTML y CSS
header = """
<div style="display: flex; align-items: center;">
    <img src="https://is2-ssl.mzstatic.com/image/thumb/Purple122/v4/8b/1b/81/8b1b81e6-93b2-d199-43af-939c64bd5238/source/512x512bb.jpg" alt="Certus Logo" style="width: 100px; margin-right: 20px;">
    <div>
        <h1 style="color: #16265a; margin: 0;">PROYECTO PYTHON PARA LA CIENCIA DE DATOS</h1>
        <h3 style="color: #16265a; margin: 0;">Programa: Data Science & Machine Learning  - Toulouse Lautrec</h3>
        <h3 style="color: #16265a; margin: 0;">Elaborado por: Ing. Diego Armando Vasquez Chavez</h3>
        <p style="color: #16265a; margin: 0;">CIP 337613</p>
    </div>
</div>


"""

# Crear la interfaz Gradio
iface = gr.Interface(
    fn=predict,
    inputs=[
        gr.Dropdown(choices=[
            "AREQUIPA", "ATE", "CALLAO", "CHICLAYO", "NORTE",
            "SAN JUAN DE LURIGANCHO", "SURCO", "VILLA EL SALVADOR", "VIRTUAL"
        ], label="SEDE ESTUDIANTE"),
        gr.Dropdown(choices=[
            "Administracin de Empresas",
            "Administracin de Empresas y Gestin de Recursos Humanos",
            "Administracin de Negocios Bancarios y Financieros",
            "Administracin de Negocios Bancarios, Financieros y Banca Digital",
            "Administracin de Negocios Internacionales",
            "Administracin de Sistemas para la Transformacin Digital",
            "Administracin y Gestin Comercial",
            "Contabilidad",
            "Contabilidad y Tributacin",
            "Diseo Grfico",
            "Diseo y Desarrollo de Software",
            "Marketing",
            "Marketing y Gestin de medios digitales",
            "Publicidad"
        ], label="PROGRAMA"),
        gr.Dropdown(choices=[
            "CICLO 01", "CICLO 02", "CICLO 03", "CICLO 04", "CICLO 05", "CICLO 06"
        ], label="CICLO"),
        gr.Dropdown(choices=["MAANA", "DIURNO", "TARDE", "NOCHE"], label="TURNO"),
        gr.Slider(minimum=0, maximum=20, step=1, label="Nota de la Evaluacin 1")
    ],
    outputs=gr.HTML(label="Probabilidades de Aprobacin, Desaprobacin e Inasistencia"),
    title="PREDICCIN DE APROBACIN, DESAPROBACIN E INASISTENCIA",
    description=header,
    css="""
        .gradio-container h1 {
            color: #16265a !important;
        }
        button[type="submit"] {
            background-color: #16265a !important;
            color: white !important;
        }
        .output-html div {
            font-size: 18px !important;
            font-weight: bold !important;
            color: #16265a !important;
        }
        .output-html ul {
            font-size: 16px !important;
            font-weight: bold !important;
            color: #16265a !important;
        }
        button[type="submit"]::after {
            content: 'Enviar';
        }
        .gradio-container button {
            text-transform: none !important;
        }
        .output-html div span {
            font-weight: bold !important;
        }
    """
)

# Ejecutar la interfaz
if __name__ == "__main__":
    iface.launch()
```
